```python
df=df[df["Reviews"].str.isnumeric()]
df["Reviews"]=df["Reviews"].astype(int)
df
```

```python
df_copy=df.copy()  


df_copy["Size"].unique()
```

```python
df_copy["Size"]=df_copy["Size"].str.replace("M","000")
df_copy["Size"]=df_copy["Size"].str.replace(".","")
df_copy["Size"]=df_copy["Size"].str.replace("K","")
df_copy["Size"]=df_copy["Size"].str.replace("k","")
df_copy=df_copy[df_copy["Size"]!='Varies with device']
df_copy["Size"]=df_copy["Size"].astype(str)
df_copy
```

```python
df_copy["Installs"].unique()
```

```python
df_copy["Installs"]=df_copy["Installs"].str.replace("+","")
df_copy["Installs"]=df_copy["Installs"].str.replace(",","")
```

```python
df_copy["Price"]=df_copy["Price"].str.replace("$","")
df_copy["Price"].unique()
```

```python
df_copy["Installs"]=df_copy["Installs"].astype(float)
df_copy["Price"]=df_copy["Price"].astype(float)
df_copy["Size"]=df_copy["Size"].astype(float)
df_copy
```

```python
df_copy["Last Updated"]=pd.to_datetime(df_copy["Last Updated"])
df_copy.to_csv("play_store_data.csv")
```

```python
numeric_features = [feature for feature in df_copy.columns if df_copy[feature].dtype != 'O']
categorical_features = [feature for feature in df_copy.columns if df_copy[feature].dtype == 'O']

# print columns
print('We have {} numerical features : {}'.format(len(numeric_features), numeric_features))
print('\nWe have {} categorical features : {}'.format(len(categorical_features), categorical_features))
```

```python
## Proportion of count data on categorical columns
for col in categorical_features:
    print(df[col].value_counts(normalize=True)*100)
    print('---------------------------')
```

```python
## Proportion of count data on numerical columns

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
plt.figure(figsize=(15, 15))
plt.suptitle('Univariate Analysis of Numerical Features', fontsize=20, fontweight='bold', alpha=0.8, y=1.)

for i in range(0, len(numeric_features)):
    plt.subplot(5, 3, i+1)
    sns.kdeplot(x=df_copy[numeric_features[i]],shade=True, color='b')
    plt.xlabel(numeric_features[i])
    plt.tight_layout()
```

```python
# categorical columns
plt.figure(figsize=(20, 15))
plt.suptitle('Univariate Analysis of Categorical Features', fontsize=20, fontweight='bold', alpha=0.8, y=1.)
category = [ 'Type', 'Content Rating']
for i in range(0, len(category)):
    plt.subplot(2, 2, i+1)
    sns.countplot(x=df[category[i]],palette="Set2")
    plt.xlabel(category[i])
    plt.xticks(rotation=45)
    plt.tight_layout() 
```

```python
## Top 10 App Categories
category = pd.DataFrame(df_copy['Category'].value_counts())        #Dataframe of apps on the basis of category
category.rename(columns = {'Category':'Count'},inplace=True)
category
```

```python
category_df = category[:10].reset_index()
category_df.columns = ['Category', 'Count']

plt.figure(figsize=(15,6))
sns.barplot(x='Category', y='Count', data=category_df, palette='hls')
plt.title('Top 10 App categories')
plt.xticks(rotation=90)
plt.show()
```

```python
#which catagory has highest number of installations
# Step 1: Group and sum installs by category
category = df_copy.groupby('Category')['Installs'].sum().sort_values(ascending=False)

# Step 2: Convert to DataFrame for Seaborn
category_df = category.reset_index()
category_df.columns = ['Category', 'Total_Installs']

# Step 3: Plot the top 10 categories
plt.figure(figsize=(15,6))
sns.barplot(x='Category', y='Total_Installs', data=category_df[:10], palette='hls')
plt.title('Top 10 App Categories by Total Installs')
plt.xticks(rotation=90)
plt.ylabel('Total Installations')
plt.tight_layout()
plt.show()
```
